```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 49, 67, 75, 90]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.show()
```

```python
# Bar plot 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

data=pd.DataFrame({
    "Class" : ['A','B','C','D','E'],
    "Students" : [57,45,33,55,12]
})
sns.barplot(x='Class',y="Students",data=data)
plt.title("Bar plot")
plt.show()
```

```python
# histogram
import numpy as np 

data=np.random.randn(100)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
#KDE plot

sns.kdeplot(data,shade= True)
plt.title("KDE Plot")
plt.show()
```

```python
# histogram with KDE

sns.histplot(data,bins=10,kde= True)
plt.title("Histogram with KDE")
plt.show()
```

```python
# box plot

sns.boxplot(y=data)
plt.title("Box PLot")
plt.show()
```

```python
#violin plot
sns.violinplot(y=data)
plt.title("violin plot")
plt.show()
```

```python
# scatter plot
x =[1,2,3,4,5]
y =[5,4,7,10,8]
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

```python
# seaborn 
sns.scatterplot(x=x)(y=y)
plt.title("scatterplot")
plt.show()
```
